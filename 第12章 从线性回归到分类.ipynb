### 12.1线性回归算法

#### 1.导入Advertising数据集

```python
#导入数据
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
import pandas as pd
data = pd.read_csv('Advertising.csv')
data.head()
```

```python
#显示TV与sales散点图
plt.scatter(data['TV'],data['sales'],color='black')
plt.xlabel('TV Ad')  
plt.ylabel('sales') 
plt.show()
```

```python
#显示radio与sales散点图
plt.scatter(data['radio'],data['sales'],color='black')
plt.xlabel('radio Ad')  
plt.ylabel('sales') 
plt.show()
```

```python
#显示newspaper与sales散点图
plt.scatter(data['newspaper'],data['sales'],color='black')
plt.xlabel('newspaper Ad')  
plt.ylabel('sales') 
plt.show()
```

#### 2.训练线性回归模型

```python
# 定义与训练线性回归模型
X = data[['TV', 'radio', 'newspaper']]
y = data['sales']
linreg = LinearRegression()
linreg.fit(X, y)
```

```python
##显示各拟合项系数
print(linreg.intercept_)
print(linreg.coef_)
```

#### 3.模型预测与检验

```python
#计算评价指标R方
y_pred = linreg.predict(X)
r=y_pred-y.mean()
r=r**2
RSS=r.sum()
r=y-y.mean()
r=r**2
TSS=r.sum()
R_squre=RSS/TSS
R_squre
```

### 12.2 逻辑回归分类算法

```python
#在iris数据集上训练逻辑回归模型
from sklearn.linear_model import LogisticRegression 
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
#导入数据
iris = load_iris()
#定义模型与训练
logreg = LogisticRegression() 
logreg.fit(iris.data,iris.target)
#预测
predicted = logreg.predict(iris.data)
#评估模型精度
accuracy_score(iris.target,predicted)
```

```python
#查看各类型预测的概率值
pred_prob=logreg.predict_proba(iris.data)[0:5,]
pred_prob
```
